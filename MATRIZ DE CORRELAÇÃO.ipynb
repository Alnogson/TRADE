```python
import pandas as pd
import yfinance as yf

# Definindo os tickers das commodities no Yahoo Finance
cafe_ticker = 'KC=F'   # Café
soja_ticker = 'SOYB'   # Soja
etanol_ticker = 'ETH=F'  # Etanol
ouro_ticker = 'GC=F'   # Ouro

# Obtendo os dados históricos utilizando yfinance
cafe_data = yf.download(cafe_ticker, start='2017-01-01', end='2022-05-15')[['Close']].rename(columns={'Close': 'Cafe'})
soja_data = yf.download(soja_ticker, start='2017-01-01', end='2022-05-15')[['Close']].rename(columns={'Close': 'Soja'})
etanol_data = yf.download(etanol_ticker, start='2017-01-01', end='2022-05-15')[['Close']].rename(columns={'Close': 'Etanol'})
ouro_data = yf.download(ouro_ticker, start='2017-01-01', end='2022-05-15')[['Close']].rename(columns={'Close': 'Ouro'})

# Mesclando os dados de café e soja
interm = pd.merge(cafe_data, soja_data, how='inner', on='Date')

# Mesclando com os dados de etanol
commodities = pd.merge(interm, etanol_data, how='inner', on='Date')

# Mesclando com os dados de ouro
commodities = pd.merge(commodities, ouro_data, how='inner', on='Date')

# Formatando os valores para no máximo 2 casas decimais
commodities = commodities.round(2)
```

## Juntando tudo num dataframe só

```python
# Dados de ações
data.index = pd.to_datetime(data.index)
# Dados de índices
indices.index = pd.to_datetime(indices.index)
# Dados de commodities
commodities.index = pd.to_datetime(commodities.index)
# Dados de criptomoedas
cripto.index = pd.to_datetime(cripto.index)
# Cotação do dólar
dolar.index = pd.to_datetime(dolar.index)
```

```python
dfs = [data, indices, commodities, cripto, dolar]
```

```python
from functools import reduce

df_final = reduce(lambda left, right: pd.merge(left, right, how = 'inner', on = 'Date'), dfs)
```

# Matriz de Correlação

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (16,8))

sns.heatmap(df_final.corr(),
           vmin = -1, vmax = 1, annot = True, cmap = 'BrBG')
```

A correlação é uma medida estatística que indica a relação entre duas variáveis. Ela varia de -1 a 1:

Correlação positiva (0 a +1): Indica que as variáveis tendem a se mover na mesma direção. Quanto mais próximo de +1, mais forte é essa relação positiva.

Correlação negativa (0 a -1): Indica que as variáveis tendem a se mover em direções opostas. Quanto mais próximo de -1, mais forte é essa relação negativa.
