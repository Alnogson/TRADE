# CALCULANDO A VOLATILIDADE DE 2 ATIVOS

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Definindo os tickers do Ibovespa e do S&P 500
ibov_ticker = '^BVSP'
sp500_ticker = '^GSPC'

# Baixando os dados histricos do Yahoo Finance
ibov_data = yf.download(ibov_ticker, start='2020-01-01', end='2023-12-31')
sp500_data = yf.download(sp500_ticker, start='2020-01-01', end='2023-12-31')

# Calculando o log retorno dirio
ibov_data['Log Return'] = np.log(ibov_data['Close'] / ibov_data['Close'].shift(1))
sp500_data['Log Return'] = np.log(sp500_data['Close'] / sp500_data['Close'].shift(1))

# Calculando a volatilidade (VIX) como o desvio padro anualizado do log retorno
ibov_volatility = ibov_data['Log Return'].std() * np.sqrt(252)
sp500_volatility = sp500_data['Log Return'].std() * np.sqrt(252)

# Plotando o grfico de log retornos
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(ibov_data['Log Return'], label='Ibovespa Log Return')
plt.title('Log Returns of Ibovespa and S&P 500')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(sp500_data['Log Return'], label='S&P 500 Log Return')
plt.xlabel('Date')
plt.legend()

plt.tight_layout()
plt.show()

# Imprimindo os ndices de volatilidade
print(f'Ibovespa Volatility (VIX): {ibov_volatility:.2%}')
print(f'S&P 500 Volatility (VIX): {sp500_volatility:.2%}')
```
