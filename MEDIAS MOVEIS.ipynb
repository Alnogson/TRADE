# MÉDIAS MÓVEIS -> EMA(9) | SMA(21)(50)(200)

## IBOVESPA

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Importando dados do índice Bovespa
inicio = '2020-01-01'
fim = '2024-11-05'

bovespa = yf.download('^BVSP', start=inicio, end=fim)

# Criação e visualização das médias móveis
ema9 = bovespa['Adj Close'].ewm(span=9, adjust=False).mean()  # Média Exponencial Móvel (9)
sma21 = bovespa['Adj Close'].rolling(21).mean()  # Média Aritmética Móvel (21)
sma50 = bovespa['Adj Close'].rolling(50).mean()  # Média Aritmética Móvel (50)
sma200 = bovespa['Adj Close'].rolling(200).mean()  # Média Aritmética Móvel (200)

bovespa['EMA9'] = ema9
bovespa['SMA21'] = sma21
bovespa['SMA50'] = sma50
bovespa['SMA200'] = sma200

# Plotagem das médias móveis
plt.figure(figsize=(10, 6))
plt.plot(bovespa['Adj Close'], label='Preço de Fechamento', color='purple')
plt.plot(bovespa['EMA9'], label='EMA (9)', color='green')
plt.plot(bovespa['SMA21'], label='SMA (21)', color='blue')
plt.plot(bovespa['SMA50'], label='SMA (50)', color='yellow')
plt.plot(bovespa['SMA200'], label='SMA (200)', color='black')

plt.title('Médias Móveis para Índice Bovespa (^BVSP)')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.legend()
plt.show()
```
